Question : Write a program to find out whether a student has passed or failed. If it requires a total of 40% & at least 33% in each subject to pass.Assume 3 subjects & take marks as an input from the user.

```python
a = int(input("Enter the marks of first subject: "))
b = int(input("Enter the marks of second subject: "))
c = int(input("Enter the marks of third subject: "))

if((a or b or c)<33 ):
  print("You are fail.")

elif((a+b+c)/3<40):
  print("You are fail.")

else:
  print("Congratulation! you have passed the exam.")
```

Question : A spam comment is defined as text containing following keywords:
"Make a lot of money", "buy now", "click this", "subscribe this".Write a program to detect these spams.

```python
c1 = "Make a lot of money"
c2 ="buy now"
c3 = "click this"
c4 = "subscribe this"

comment = input("Enter your comment here : ")

if( (c1 in comment) or (c2 in comment) or (c3 in comment) or (c4 in comment)):
  print("This is a spam comment.")

else:
  print(comment)
```

Question : write a program to calculate the grade of a student from his marks from the following scheme:
90-100 = A+,
80-90 = A,
70-80 = B,
60-70 = C,
50-60 = D,
<50 = F.

```python
marks = int(input("Enter your marks here: "))

if(marks<=100 and marks>90):
  print("Your grade is A+ .")

elif(marks<=90 and marks>80):
  print("Your grade is A .")

elif(marks<=80 and marks>70):
  print("Your grade is B .")

elif(marks<=70 and marks>60):
  print("Your grade is C .")

elif(marks<=60 and marks>50):
  print("Your grade is D .")

elif(marks<50 and marks>=0):
  print("Your grade is F .")
```

Quesion : Write a program to find out whether the given post is talking about "Rohit" or not.

```python
post = input("Enter your post here: ")

if("Rohit".lower() in post.lower()):
  print("The post is talking about Rohit.")

else:
  print("The post is not talking about Rohit.")
```
