Question : Write a class "Calculator" capable of finding square , cube and square root of a number.

````python
num = int(input("Enter the number here: "))

class Calculator:
  def __init__(self,num):
    self.number = num

  def square(self):
    print(f'The square of {self.number} is {self.number **2}')

  def squareRoot(self):
    print(f'The squareroot of {self.number} is {self.number **0.5}')

  def cube(self):
    print(f'The cube of {self.number} is {self.number **3}')

a = Calculator(num)
a.square()
a.squareRoot()
a.cube()
````

Question : Add a static method in problem 2 to greet the use with "Hello".

````python
num = int(input("Enter the number here: "))

class Calculator:
  def __init__(self,num):
    self.number = num

  def square(self):
    print(f'The square of {self.number} is {self.number **2}')

  def squareRoot(self):
    print(f'The squareroot of {self.number} is {self.number **0.5}')

  def cube(self):
    print(f'The cube of {self.number} is {self.number **3}')

  @staticmethod  # because we don't haave to pass any value or self here.
  def greet():
    print("hello there")


a = Calculator(num)
a.greet()
a.square()
a.squareRoot()
a.cube()
````

Question : Create a class (2-D vector) and use it to create another class representing a 3-D vector.

````python
class TwoDVector:
  def __init__(self,i,j):
    self.i = i
    self.j = j

  def show(self):
    print(f"The vecotr is {self.i}i + {self.j}j")

class ThreeDVector(TwoDVector):
  def __init__(self,i,j,k):
    super().__init__(i,j)
    self.k = k

  def show(self):
    print(f"The vecotr is {self.i}i + {self.j}j + {self.k}k")


v2d = TwoDVector(10,20)
v2d.show()
v3d = ThreeDVector(10,20,30)
v3d.show()
````

Question : Create a class 'Employee' and add salary and increment properties to it.
Write a method 'salaryAfterIncrement method with a @property decorator with a setter which changes the value of increment based on the salary.

````python
class Employee():
  salary = 2500
  increment = 20

  @property
  def salaryAfterIncrement(self):
    return self.salary + self.salary*self.increment/100

  @salaryAfterIncrement.setter
  def salaryAfterIncrement(self,sai):
    self.increment = sai/self.salary

e = Employee()
print(e.salaryAfterIncrement)
print(e.increment)
````

Question :  Write a class "Complex" to represent complex numbers, along with overloaded operators '+' and '*' which adds and multiplies them.

````python
class Complex:
  def __init__(self,r,i):
    self.real = r
    self.imaginary = i

  def __add__(self,c):
    return Complex(self.real + c.real,self.imaginary + c.imaginary)

  def __mul__(self,c):
    mulReal = self.real*c.real - self.imaginary*c.imaginary
    mulImg = self.real*c.imaginary + self.imaginary*c.real
    return Complex(mulReal,mulImg)

  def __str__(self):
    if self.imaginary<0:
      return f"{self.real} - {-self.imaginary}i"
    else:
      return f"{self.real} + {self.imaginary}i"

c1 = Complex(1,4)
c2 = Complex(3,5)
print(c1 + c2)
print(c1 * c2)
````

Question : Write a class vector representing a vector of n dimensions. Overload + the * and operator which calculates the sum and the dot(.) product of them.

````python
class vector:
  def __init__(self,x,y,z):
    self.x = x
    self.y = y
    self.z = z

  def __str__(self):
    return f"vector({self.x},{self.y},{self.z})"

  def __add__(self,vec2):
      result = vector(self.x + vec2.x, self.y + vec2.y, self.z + vec2.z)
      return result

  def __mul__(self,vec2):
      result = (self.x * vec2.x + self.y * vec2.y + self.z * vec2.z)
      return result

v1 = vector(1,2,3)
v2 = vector(4,5,6)
v3 = vector(7,8,9)

print(v1 + v2)
print(v1 * v2)
print("")
print(v1 + v3)
print(v1 * v3)
````
