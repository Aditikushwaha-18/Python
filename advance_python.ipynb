Question 1. Write a program to open three files 1.txt, 2.txt and 3.txt if any these files are not present, a message without exiting the program must be printed prompting the same.

Question 2. Write a program to print third, fifth and seventh element from a list using enumerate function.

```python
mylist = [10, 20, 30, 40, 50, 60, 70, 80, 90]

for index, value in enumerate(mylist):
    if (index == 2 or index == 4 or index == 6) :
        print(value)

```

Question 3. Write a list comprehension to print a list which contains the multiplication table of a user entered number.

```python
n = int(input("Enter a number: "))

table = [n*i for i in range(1,11)]
print(table)
```

Question 5. Store the multiplication tables generated in problem 3 in a file named Tables.txt.

```python
n = int(input("Enter a number: "))

table = [n*i for i in range(1,11)]
print(table)

with open('Tables.txt', 'w') as f:
    f.write(str(table))
```

Question 7. A list contains the multiplication table of 7. write a program to convert it to vertical string of same numbers.

```python
table = [str(7*i) for i in range(1,11)]

s = "\n".join(table)
print(s)
```

Question 9. Write a program to find the maximum of the numbers in a list using the reduce function.

```python
from functools import reduce

l = [23, 54, 76, 98, 29, 91, 82, 18, 1932, 4, 34, 293]

def greater(a,b):
  if(a>b):
    return a
  return b

k = (reduce(greater,l))
print(k)
```
