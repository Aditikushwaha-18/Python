Question : Write a program to print multiplication table using for loop.

Question : Wrrite a program to find whether the given number is prime or not.

````python
n = int(input("Enter the number:"))
for i in range(2,n):
  if (n%i == 0):
    print("Not Prime")
    break
else:
  print("Prime")
````

Question : Write a program to find the sum of n natural numbers using a while loop.

````python
n = int(input("Enter the number:"))
sum = 0
i=1
while (i<=n):
  sum+=i
  i+=1
print(sum)
````

Question : Write a program to find out factorial of the given number using for loop.

````python
n = int(input("Enter the number:"))
fact = 1
for i in range(1,n+1):
  fact*=i

print(fact)
````

Question : Write a program to print the following star pattern.

```
***
* *
***
```
for n=3.


````python
n = int(input("Enter the number:"))

for  i in range(1,n+1):
  if(i==1 or i==n):
    print("*"*n)
  else:
    print("*",end="")
    print(" "*(n-2),end="")
    print("*")
````

Question : Write a program to print multiplication table of n using for loops in reverse order.

````python
n = int(input("Enter the number:"))
for i in range(1,11):
    print(f"{n} X {11-i} = {n*(11-i)}")
````
