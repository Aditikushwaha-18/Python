Question : Write a program to create dictionary of hindi words with values as their english translation . Provide user with an option to look it up!

```python
words = {
    "madat":"help",
    "chappal":"slippers",
    "kursi":"chair",
    "rumaal":"hankerchief",
}

word = input("Enter the hindi word : ")

print(words[word])
```

Question : What will be the lenght of the following set s :
s = set()
s.add(20)
s.add(20.0)
s.add('20')  # length of s after these operations?


```python
s = set()
s.add(20)
s.add(20.0)
s.add('20')

print(len(s))   # the ans is 2 because in python if the values are numerically same it is considered as same , data type is not considered here.
```

Question : Create an empty  dictionary . Allow 4 friends to enter thier favourite language as a value and use key as their names. Assume that the names are unique.

```python
language = {
    "Ram" :   "",
    "Shyam" : "",
    "Mohan" : "",
    "Sohan" : "",
}
l1 = input("Enter your favourite language Ram : ")
language["Ram"]=l1
l2 = input("Enter your favourite language Shyam : ")
language["Shyam"]=l2
l3 = input("Enter your favourite language Mohan : ")
language["Mohan"]=l3
l4 = input("Enter your favourite language Sohan : ")
language["Sohan"]=l4

print(language)
```
